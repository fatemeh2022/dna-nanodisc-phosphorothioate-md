# only DMPC

```python
# Nanodisc Lipid Estimator (Å units)

import numpy as np

# === INPUT ===
diameter_A = 115.9   # Measured from PyMOL in Å (11.59 nm = 115.9 Å)
APL = 64.1           # Area per lipid in Å² (typical for DMPC)

# === CALCULATION ===
radius = diameter_A / 2.0              # in Å
area_one_leaflet = np.pi * radius**2   # Å²
lipids_per_leaflet = area_one_leaflet / APL
total_lipids = int(2 * lipids_per_leaflet)

# === OUTPUT ===
print("Formulas:")
print("  Circumference: C = 2πr")
print("  Area:          A = πr²\n")
print(f"Measured diameter: {diameter_A:.1f} Å")
print(f"Calculated radius: {radius:.2f} Å")
print(f"Area per leaflet: {area_one_leaflet:.2f} Å²")
print("area per lipid (APL): {:.1f} Å²".format(APL))
print(f"Estimated lipids per leaflet: {lipids_per_leaflet:.1f}")
print(f"✅ Total estimated DMPC lipids (both leaflets): {total_lipids}")

```

# DMPC/DMTAP


```python
import numpy as np

# --- Inputs ---
bp = 147
rise_per_bp_nm = 0.34
DNA_thickness_nm = 2.0
belt_thickness_nm = 0.0

APL_DMPC = 64.1     # Å^2
APL_DMTAP = 60.0    # Å^2
DMTAP_fraction = 0.15
DMPC_fraction = 0.85

pi = np.pi

# --- Geometry ---
C_nm = bp * rise_per_bp_nm
dna_diameter_nm = C_nm / pi
inner_diameter_nm = dna_diameter_nm - DNA_thickness_nm - belt_thickness_nm
inner_radius_A = (inner_diameter_nm * 10.0) / 2.0

# --- Effective APL ---
APL_eff = (DMPC_fraction * APL_DMPC + DMTAP_fraction * APL_DMTAP)



APL_eff = (
    0.85 * APL_DMPC   # contribution from DMPC
  + DMTAP_fraction * APL_DMTAP  # contribution from DMTAP
)

# --- Areas & counts ---
area_one_leaflet_A2 = pi * inner_radius_A**2
lipids_per_leaflet = area_one_leaflet_A2 / APL_eff
total_lipids = int(round(2 * lipids_per_leaflet))

# --- Composition ---
n_dmtap = int(round(total_lipids * DMTAP_fraction))
n_dmpc = total_lipids - n_dmtap

# --- Output ---
print("Formulas:")
print("  Effective APL = x_DMPC * APL_DMPC + x_DMTAP * APL_DMTAP\n")

print(f"DNA circle diameter: {dna_diameter_nm:.2f} nm")
print(f"Inner diameter: {inner_diameter_nm:.2f} nm ({inner_diameter_nm*10:.1f} Å)")
print(f"Effective APL: {APL_eff:.2f} Å^2")
print(f"Area per leaflet: {area_one_leaflet_A2:.1f} Å^2")
print(f"Estimated lipids per leaflet: {lipids_per_leaflet:.1f}")

print("\nComposition:")
print(f"Total lipids (both leaflets): {total_lipids}")
print(f"DMTAP (15%): {n_dmtap}")
print(f"DMPC  (85%): {n_dmpc}")
```
