```python
import numpy as np
import matplotlib.pyplot as plt

def load_xvg(filename):
    data = np.loadtxt(filename, comments=["@", "#"])
    return data[:, 0], data[:, 1]  # time, value


# Load files
t_time, temp = load_xvg("Temperature.xvg")
p_time, pressure = load_xvg("Pressure.xvg")
d_time, density = load_xvg("Density.xvg")


# Plot Temperature
plt.figure()
plt.plot(t_time, temp)
plt.xlabel("Time (ps)")
plt.ylabel("Temperature (K)")
plt.title("Temperature vs Time")
plt.savefig("temperature.png", dpi=300, bbox_inches="tight")  # Save as PNG
print("Average Temperature:", temp.mean())
print("Std deviation:", temp.std())
plt.show()


# Plot Pressure
plt.figure()
plt.plot(p_time, pressure)
plt.xlabel("Time (ps)")
plt.ylabel("Pressure (bar)")
plt.title("Pressure vs Time")
plt.savefig("pressure.png", dpi=300, bbox_inches="tight")  # Save as PNG
plt.show()
print("Average pressure:", pressure.mean())
print("Std deviation:", pressure.std())

# Plot Density
plt.figure()
plt.plot(d_time, density)
plt.xlabel("Time (ps)")
plt.ylabel("Density (kg/m^3)")
plt.title("Density vs Time")
plt.savefig("density.png", dpi=300, bbox_inches="tight")  # Save as PNG
plt.show()
```
