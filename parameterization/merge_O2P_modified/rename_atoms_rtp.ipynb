```python
import re

# Read the file content
with open("DA_.rtp", "r") as file:
    text = file.read()

# Replacement dictionary
replacements = {
    'O2':"O5'",
    'C1':"C5'",
    'C3':"C4'",
    'C5':"C3'",
    'C7':"C2'",
    'C8':"C1'",
    'O4':"O4'",
    'N18':'N9',
    'C17':'C8',
    'N16':'N7',
    'C13':'C5',
    'C14':'C6',
    'N9':'N1',
    'C10':'C2',
    'N11':'N3',
    'C12':'C4',
    'N15':'N6',
    'H30':"HO5'",
    'H25':"H5'1",
    'H29':"H5'2",
    'H24':"H4'",
    'H23':"H3'",
    'H31':"HO3'",
    'H22':"H2'1",
    'H28':"H2'2",
    'H21':"H1'",
    'H20':'H8',
    'H19':'H2',
    'H26':'H61',
    'H27':'H62'  

}

# Replace with a regex that matches atom names surrounded by spaces, tabs, or line breaks
for old_word, new_word in replacements.items():
    pattern = rf'(?<=\s){old_word}(?=\s|$)'  # match only if surrounded by spaces or end of line
    text = re.sub(pattern, new_word, text)

# Save the modified file
with open("DA_rename.rtp", "w") as file:
    file.write(text)

print("DA changes atom names in atom and bonds sections.")
```

```python
# Read the first modified file
with open("DA_rename.rtp", "r") as f1:
    dti_text = f1.read()

# Read the second modified file
with open("bak_rename.rtp", "r") as f2:
    linker_text = f2.read()

# Combine them with a separator (optional)
merged_text = dti_text.strip() + "\n\n; ====== LINKER SECTION BELOW ======\n\n" + linker_text.strip()

# Save the merged file
with open("merged_output.rtp", "w") as fout:
    fout.write(merged_text)

print("✅ Done! Created merged_output.rtp with both modified contents.")
```
