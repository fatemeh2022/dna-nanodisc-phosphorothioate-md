```python
import re

# Read the file content
with open("DC_.rtp", "r") as file:
    text = file.read()

# Replacement dictionary
replacements = {
    "C1": "C5'",
    "O2": ";RO5'",
    "C3": "C4'",
    "O4": "O4'",
    "C5": "C3'",
    "O6": ";RO3'",
    "C7": "C2'",
    "C8": "C1'",
    "N9": "N1",
    "C10": "C2",
    "O11": "O2",
    "N12": "N3",
    "C13": "C4",
    "N14": "N4",
    "C15": "C5",
    "C16": "C6",
    "H17": "H5",
    "H18": "H6",
    "H19": "H1'",
    "H20": "H2'1",
    "H21": "H3'",
    "H22": "H4'",
    "H23": "H41",
    "H24": "H5'1",
    "H25": "H42",
    "H26": "H2'2",
    "H27": "H5'2",
    "H28": ";RHO5'",
    "H29": ";RHO3'"
}

# Replace with a regex that matches atom names surrounded by spaces, tabs, or line breaks
for old_word, new_word in replacements.items():
    pattern = rf'(?<=\s){old_word}(?=\s|$)'  # match only if surrounded by spaces or end of line
    text = re.sub(pattern, new_word, text)

# Save the modified file
with open("DC_rename.rtp", "w") as file:
    file.write(text)

print("DC changes atom names in atom and bonds sections.")
```
